```python
from bertopic import BERTopic
```

```python
import pandas as pd
```

```python
reviews_df=pd.read_csv('all_review_processed.csv',engine='python',sep=',', quotechar='"', error_bad_lines=False)

review_text=reviews_df.Text.tolist()
```

```python
reviews_text = review_text[:50000]
```

```python
reviews_text
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(ngram_range=(1, 2), stop_words="english")
topic_model = BERTopic(vectorizer_model=vectorizer_model,calculate_probabilities=True)
```

```python
topics, probabilities = topic_model.fit_transform(reviews_text)
```

```python
model = BERTopic(calculate_probabilities=True)
```

```python
topics, probabilities = model.fit_transform(reviews_text)
```

```python
model.get_topic_freq().head()
```

```python
model.save("my_model")
```
